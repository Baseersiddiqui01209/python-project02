```python
dv.director.fillna(dv.director.mode()[0],inplace=True)
dv.cast.fillna(dv.cast.mode()[0],inplace=True)
dv.country.fillna(dv.country.mode()[0],inplace=True)
dv.date_added.fillna(dv.date_added.mode()[0],inplace=True)
dv.rating.fillna(dv.rating.mode()[0],inplace=True)
dv.duration.fillna(dv.duration.mode()[0],inplace=True)
```

```python
dv.head()
```

```python
dv.isna().sum()
```

```python
# Find out which types of genre has most movies and TV Shows

most_common_genre = dv['listed_in'].str.split(', ').explode().mode()[0]

print(f"The most common genre is: {most_common_genre}")
```

```python
#Get the top 10 countries contributing the most content
top_countries = dv['country'].value_counts().nlargest(10)

print(top_countries)
```

```python
# Create the barplot
sns.barplot(x='Country', y='Number of Movies', data=top_countries, palette='Blues')

# Add title and labels
plt.title('Top 10 Countries with Most Movies on Netflix', fontsize=12)
plt.xlabel('Country', fontsize=10)
plt.ylabel('Number of Movies', fontsize=12)
plt.show()
```
